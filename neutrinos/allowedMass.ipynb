```python
from massConstraints import find_allowed_lightest_masses, sum_mass_constraint

import numpy as np
import matplotlib.pyplot as plt

# Compute allowed lightest masses (normal ordering)
m1, m2, m3, Mtot = find_allowed_lightest_masses(ordering="normal")
```

```python
# Plot the three masses and the total
plt.figure(figsize=(8,5))
plt.plot(m1, Mtot, label=r'$M_{\rm tot}$ (eV)', lw=2, ls='--', color='black')
plt.axhline(sum_mass_constraint, color='red', ls=':', label=fr'$\sum m_\nu$ constraint = {sum_mass_constraint} eV')
plt.xlabel('Lightest Mass (eV)')
plt.ylabel('Mass (eV)')
plt.title('Total Neutrino Mass vs Lightest Mass (normal ordering)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot the three masses and the total
plt.figure(figsize=(12,5))
plt.plot(m1, m1, label='m1 (eV)', lw=2)
plt.plot(m1, m2, label='m2 (eV)', lw=2)
plt.plot(m1, m3, label='m3 (eV)', lw=2)
plt.plot(m1, Mtot, label='Mtot (eV)', lw=2, ls='--', color='black')
plt.axhline(sum_mass_constraint, color='red', ls=':', label=f'sum constraint = {sum_mass_constraint} eV')
plt.xlabel('Lightest mass (eV)')
plt.ylabel('Mass (eV)')
plt.title('Neutrino masses vs lightest mass (normal ordering)')
# place legend outside plot on the right
plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1), borderaxespad=0)
# leave room on the right for the legend
plt.tight_layout(rect=[0, 0, 0.78, 1])
plt.grid(True)
plt.show()
```
